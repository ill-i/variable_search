```python
# Попытка исправить код для настройки штриховки
import matplotlib.pyplot as plt

# Данные
total_files = 1563
calibration_files = 160
photometric_files = 1403
usable_files = 1316

# Калибровочные кадры
bad_calibration_files = 60
good_calibration_files = 100
good_calibration_no_coords = 46
good_calibration_with_coords = 54

# Фотометрические кадры объектов
successful_astrometry_files = 1216
failed_astrometry_files = 187

# Готовим данные для графика
labels = [
    'Калибровочные кадры без экспозиций',
    'Калибровочные кадры без координат',
    'Калибровочные кадры с полными данными',
    'Фотометрические кадры с астрометрией',
    'Фотометрические кадры без астрометрии'
]

sizes = [
    bad_calibration_files,
    good_calibration_no_coords,
    good_calibration_with_coords,
    successful_astrometry_files,
    failed_astrometry_files
]

# Цвета для сегментов графика
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0']

# Штриховки для сегментов
hatches = ['/', '\\', '|', '-', '+']

# Создаем pie chart с отдельной легендой и штриховками
plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(sizes, colors=colors, autopct='%1.1f%%', startangle=142, 
                                   textprops={'fontsize': 14, 'fontweight': 'bold', 'color': 'black'},
                                   wedgeprops={'edgecolor': 'white'})

# Добавляем штриховки к каждому сегменту
for wedge, hatch in zip(wedges, hatches):
    wedge.set_hatch(hatch)
    wedge.set_edgecolor('gray')  # установка цвета края сегмента в серый

# Настраиваем легенду
plt.legend(wedges, labels, loc="lower center", bbox_to_anchor=(0.5, -0.11), ncol=2)

#plt.title('Распределение обработанных фотометрических данных')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Отображаем график
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

# Данные
total_files = 1563
calibration_files = 160
photometric_files = 1403
usable_files = 1316

# Калибровочные кадры
bad_calibration_files = 60
good_calibration_files = 100
good_calibration_no_coords = 46
good_calibration_with_coords = 54

# Фотометрические кадры объектов
successful_astrometry_files = 1216
failed_astrometry_files = 187

# Готовим данные для графика
labels = [
    'Calibration frames without exposures',
    'Calibration frames without coordinates',
    'Calibration frames with full data',
    'Photometric frames with astrometry',
    'Photometric frames without astrometry'
]

sizes = [
    bad_calibration_files,
    good_calibration_no_coords,
    good_calibration_with_coords,
    successful_astrometry_files,
    failed_astrometry_files
]

# Цвета для сегментов графика
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0']

# Штриховки для сегментов
hatches = ['/', '\\', '|', '-', '+']

# Функция для отображения процентов внутри сегментов
def autopct_generator(pct):
    return f'{pct:.1f}%'

# Создаем pie chart с отдельной легендой и штриховками
plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(sizes, colors=colors, startangle=140, autopct=autopct_generator, 
                                   textprops={'fontsize': 14,  'color': 'black'},#'fontweight': 'bold',
                                   wedgeprops={'edgecolor': 'black', 'linewidth': 1.2, 'width': 0.4})

# Добавляем штриховки к каждому сегменту
for wedge, hatch in zip(wedges, hatches):
    wedge.set_hatch(hatch)

# Добавляем проценты внутри сегментов, сдвигая их ближе к центру
for autotext in autotexts:
    autotext.set_color('black')  # Цвет текста
    autotext.set_path_effects([path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])
    x, y = autotext.get_position()
    autotext.set_position((x*0.75, y*0.85))  # Сдвигаем проценты ближе к центру

# Настраиваем легенду
plt.legend(wedges, labels, loc="lower center", bbox_to_anchor=(0.5, -0.17), ncol=2, fontsize=14)

#plt.title('Распределение обработанных фотометрических данных')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Сохраняем график
plt.tight_layout()
plt.savefig("schmidt_stat.png", dpi=300, bbox_inches='tight')
plt.show()
```
